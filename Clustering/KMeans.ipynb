```python
import torch
from torch import nn
import numpy
import matplotlib.pyplot as plt
from pyecharts.charts import Scatter
from pyecharts import options
```

```python
origin = torch.ones((40, 2))
a = torch.normal(origin * 10, torch.tensor(2.))
b = torch.normal(origin * -10, torch.tensor(2.))
```

```python
plt.scatter(a.numpy()[:, 0], a.numpy()[:, 1],c='r')
plt.scatter(b.numpy()[:, 0], b.numpy()[:, 1],c='b')
plt.show()
```

```python
scatter = Scatter()
scatter.add_xaxis(xaxis_data=a.numpy()[:, 0].tolist())
scatter.add_yaxis(series_name="A",
                  y_axis=a.numpy()[:, 1].tolist(),
                  label_opts=options.LabelOpts(is_show=False))
scatter.add_xaxis(xaxis_data=b.numpy()[:, 0].tolist())
scatter.add_yaxis(series_name="B",
                  y_axis=b.numpy()[:, 1].tolist(),
                  label_opts=options.LabelOpts(is_show=False))
scatter.set_global_opts(
    xaxis_opts=options.AxisOpts(
        type_="value", splitline_opts=options.SplitLineOpts(is_show=True)),
    yaxis_opts=options.AxisOpts(
        type_="value", splitline_opts=options.SplitLineOpts(is_show=True)))
scatter.render_notebook()
```

```python
c = torch.cat((a, b))
centers = torch.tensor([[1., 1.], [-1., -1.]])
for step in range(1000):
    distance1 = ((c - centers[0])**2).sum(dim=1).sqrt()
    distance2 = ((c - centers[1])**2).sum(dim=1).sqrt()
    labels = [
        0 if distance1[i] < distance2[i] else 1 for i in range(c.shape[0])
    ]
    distance_sum = torch.zeros((2, 2))
    num1 = 0
    num2 = 0
    for i in range(c.shape[0]):
        if labels[i] == 0:
            distance_sum[0] = distance_sum[0].add(c[i])
            num1 += 1
        else:
            distance_sum[1] = distance_sum[1].add(c[i])
            num2 += 1
    distance_sum[0] /= num1
    distance_sum[1] /= num2
    centers = distance_sum
    print(centers)
```
