# 硬间隔SVM

也叫最大间隔分类器



超平面$w^Tx+b$来区分图中两类点

判别模型$f(w)=sign(w^Tx+b)$，$w^Tx+b>0$，则$f(w)=1$；$w^Tx+b<0$，则$f(w)=-1$，是二分类问题

最大间隔：$\max margin(w,b)$

约束：$w^Tx_i+b>0, y_i=1$；$w^Tx_i+b<0, y_i=-1$

因为$y_i$和$w^Tx_i+b$同号，所以约束可以简化为：

$y_i(w^Tx_i+b)>0$



公式中$margin$定义为距离超平面最近的点到超平面的距离

点到直线距离公式为：$distance=\frac{1}{||w||}|w^Tx_i+b|$

$maring(w,b)=\min distance(w,b,x_i)=\min \frac{1}{||w||}|w^Tx_i+b|$

代入最大间隔公式：

$\max \min \frac{1}{||w||}|w^Tx_i+b|$，约束条件$y_i(w^Tx_i+b)>0$

可以把约束条件代入公式，去掉绝对值且保持式子意义不变：

$\max \min \frac{1}{||w||}y_i(w^Tx_i+b) = \max \frac{1}{||w||} \min y_i(w^Tx_i+b)$

从约束条件可以推出一个关系：

$\exists r>0, \min y_i(w^Tx_i+b)=r$

所以可以把$r$代入其中，并且因为超平面的系数可以随意缩放，所以可以直接将$r$缩放成1，那个最大间隔就直接变成这样：

$$
\begin{cases}
\max \frac{1}{||w||} \\
s.t.\ \min y_i(w^Tx_i+b)=1 
\end{cases}
$$

再化简一下形式

$$
\begin{cases}
\min ||w|| = \min \frac{1}{2} w^Tw\\
s.t.\ y_i(w^Tx_i+b)\ge1
\end{cases}
$$

这是一个凸优化问题

# 对偶

由于这个问题的特殊结构，还可以通过拉格朗日对偶性变换到对偶变量的优化问题，即通过求解与原问题等价的对偶问题得到原始问题的最优解，这就是线性可分条件下支持向量机的对偶算法，这样做的优点在于：
1. 对偶问题往往更容易求解
2. 可以自然的引入核函数，进而推广到非线性分类问题

那什么是拉格朗日对偶性呢？简单来讲，通过给每一个约束条件加上一个拉格朗日乘子$\alpha$，定义拉格朗日函数（通过拉格朗日函数将约束条件融合到目标函数里去，从而只用一个函数表达式便能清楚的表达出我们的问题）：

$L(w,b,\alpha)=\frac{1}{2}w^Tw-\Sigma a_i(y_i(w^Tx_i+b)-1)$

令$\theta(w)=\max_{\alpha_i \ge 0} L(w,b,\alpha)$
