```python
import torch
from torch import nn
from torch import optim
from torch.nn import functional
```

```python
n_movie = 5
n_user = 4
scores = torch.tensor(
    [[4, 5, 0, 0], [5, 5, 1, 0], [4, 4, 1, 0], [0, 1, 5, 5], [0, 0, 4, 5]],
    dtype=torch.float32)

u = torch.randn((n_user, 2), requires_grad=True)
m = torch.randn((n_movie, 2), requires_grad=True)
```

```python
lr = 0.01

for step in range(10000):
    pred = m.matmul(u.transpose(0, 1))
    loss = (pred - scores).pow(2).mean()
    print("step=%d loss=%f" % (step, loss))
    loss.backward()
    with torch.no_grad():
        u -= u.grad * lr
        m -= m.grad * lr
        u.grad.zero_()
        m.grad.zero_()

print(u)
print(m)
print(m.matmul(u.transpose(0, 1)))
```

```python
class CollaborativeFilter(nn.Module):
    def __init__(self):
        super().__init__()
        self.u = nn.Parameter(torch.randn(n_user, 2))
        self.m = nn.Parameter(torch.randn(n_movie, 2))

    def forward(self):
        return self.m.matmul(self.u.transpose(0, 1))


cf = CollaborativeFilter()
optimizer = optim.SGD(cf.parameters(), lr=0.01)
```

```python
for step in range(10000):
    pred = cf()
    loss = (pred - scores).pow(2).mean()
    print("step=%d loss=%f" % (step, loss))
    loss.backward()
    optimizer.step()
    optimizer.zero_grad()
```
